## Accept Overrides

```python
df = df.with_columns(
    pl.when(pl.col("title__override").is_not_null())
    .then(pl.col('title__override').str.strip_chars())
    .otherwise(pl.col('title')).alias('title'),
    pl.when(pl.col("title__zh-hant__override").is_not_null())
    .then(pl.col('title__zh-hant__override').str.strip_chars())
    .otherwise(pl.col('title__zh-hant')).alias('title__zh-hant__override'),
    pl.when(pl.col("description__override").is_not_null())
    .then(pl.col('description__override').str.strip_chars())
    .otherwise(pl.col('description')).alias('description'),
    pl.when(pl.col("description__zh-hant__override").is_not_null())
    .then(pl.col('description__zh-hant__override').str.strip_chars())
    .otherwise(pl.col('description__zh-hant')).alias('description__zh-hant'),
    pl.when(pl.col("formattedAddress__override").is_not_null())
    .then(pl.col('formattedAddress__override').str.strip_chars())
    .otherwise(pl.col('formattedAddress')).alias('formattedAddress'),
    pl.when(pl.col("visitableAsOf__override").is_not_null())
    .then(pl.col('visitableAsOf__override').str.strip_chars())
    .otherwise(pl.col('visitableAsOf')).alias('visitableAsOf')
)
```

```python
post_merge_cols = [c for c in selected_cols if not c.endswith('__override')]
df = df[post_merge_cols]
```

```python
def get_contributor_id():
    return random.choice([BILLYID, BENID])

def gen_nano_id():
    return fastnanoid.generate(size=12)

def row_to_feature(row):
    return {
        "id" : row["id"],
        "geometry" : {
            "type": "Point",
            "coordinates": [row["longitude"], row["latitude"]]
        },
        "title" : row["title"] if row["title"] else "Undefined",
        "titleGen" : row["titleGen"] if row["titleGen"] else True,
        "description" : row["description"],
        "descriptionGen" : row["descriptionGen"] if row["descriptionGen"] else True,
        "addressProperties" : {
            "distanceFromPoint": row["distanceFromPoint"],
            "formattedAddress": row["formattedAddress"],
            "formattedAddress__zh-hant": row["formattedAddress__zh-hant"],
            "formattedAddress__zh-hans": row["formattedAddress__zh-hans"],
            "plusCode": row["plusCode"],
            "plusCode__zh-hant": row["plusCode__zh-hant"],
            "plusCode__zh-hans": row["plusCode__zh-hans"],
            "subPremise": row["subPremise"],
            "subPremise__zh-hant": row["subPremise__zh-hant"],
            "subPremise__zh-hans": row["subPremise__zh-hans"],
            "premise": row["premise"],
            "premise__zh-hant": row["premise__zh-hant"],
            "premise__zh-hans": row["premise__zh-hans"],
            "streetNumber": row["streetNumber"],
            "streetNumber__zh-hant": row["streetNumber__zh-hant"],
            "streetNumber__zh-hans": row["streetNumber__zh-hans"],
            "route": row["route"],
            "route__zh-hant": row["route__zh-hant"],
            "route__zh-hans": row["route__zh-hans"],
            "intersection": row["intersection"],
            "intersection__zh-hant": row["intersection__zh-hant"],
            "intersection__zh-hans": row["intersection__zh-hans"],
            "neighbourhood": row["neighbourhood"],
            "neighbourhood__zh-hant": row["neighbourhood__zh-hant"],
            "neighbourhood__zh-hans": row["neighbourhood__zh-hans"],
            "administrativeAreaLevel1": row["administrativeAreaLevel1"],
            "administrativeAreaLevel1__zh-hant": row["administrativeAreaLevel1__zh-hant"],
            "administrativeAreaLevel1__zh-hans": row["administrativeAreaLevel1__zh-hans"],
            "country": row["country"],
            "country__zh-hant": row["country__zh-hant"],
            "country__zh-hans": row["country__zh-hans"],
            "googlePlaceId": row["googlePlaceId"],
            "addressGeocoder": row["geocoder"],
            "addressReverseGen": row["reverseGen"],
            "addressForwardGen": row["forwardGen"]
        },
        "layerId" : LAYERID,
        "contributorId" : get_contributor_id(),
        "publisherId" : SUPERADMINID,
        "isPublished" : True,
        "isIntangible" : False,
        "isVisitable": True,
        "visitableAsOf" : row['visitableAsOf']
    }

def row_to_featureI18n(row, lang='zh-hant'):
    return {
        "featureId" : row["id"],
        "lang" : lang,
        "title" : row["title__zh-hant"] if row["title__zh-hant"] else "Undefined",
        "titleGen" : row["titleGen__zh-hant"] if lang == 'zh-hant' else True,
        "description" : row["description__zh-hant"],
        "descriptionGen" : row["descriptionGen__zh-hant"] if lang == 'zh-hant' else True,
    }

def row_to_feature_properties(row):
    properties = []
    for option_key, option in OPTIONSET.items():
        properties.append({
            "id" : gen_nano_id(),
            "featureId" : row['id'],
            "propertyId" : option['id'],
            "propertyValueId" : random.choice([v['id'] for v in option['values'].values()])
        })
    properties.append({
        "id" : gen_nano_id(),
        "featureId" : row['id'],
        "propertyId" : 'hJ4kL9mN9pQ2',
        "value" : random.choice(["1","2","3","4","5"])
    })
    for specifier, property_id in [('graphemes','tU5vW8xY1zA4'),('calligrapher','bC7dF9gH2jK5')]:
        properties.append({
            "id" : gen_nano_id(),
            "featureId" : row['id'],
            "propertyId" : property_id,
            "value" : row[specifier]
        })
    return properties
```

```python
JSON_SEED_DATA = {
    "feature" : [],
    "featureI18n" : [],
    "featureProperty" : [],
    "featurePropertyI18n" : []
}

for feature in df.rows(named=True):
    JSON_SEED_DATA['feature'].append(row_to_feature(feature))
    JSON_SEED_DATA['featureI18n'].append(row_to_featureI18n(feature))
    JSON_SEED_DATA['featureI18n'].append(row_to_featureI18n(feature, 'zh-hans'))
    JSON_SEED_DATA['featureProperty'].extend(row_to_feature_properties(feature))
```

```python
def default(o):
    if type(o) is date:
        return o.isoformat()

def jsondump(o, f):
    return json.dump(o, f, ensure_ascii=False, indent=4, default=default)

for table, data in JSON_SEED_DATA.items():
    with open(f"{JSONDEST}/{table}.json", 'w', encoding='utf-8') as f:
        jsondump(data, f)
```
